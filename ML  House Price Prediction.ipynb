***Part-2: Working with Model***

```python
# Create the target data and feature data where target data is price
x = df.drop(columns="Price")
y = df["Price"] 
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.score(x_train,y_train)*100
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import numpy as np
```

```python
y_pred = lr.predict(x_test)
```

```python
MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
R2SCORE = r2_score(y_test,y_pred)*100

print("MAE =",MAE)
print("MSE", MSE)
print("R2 Score",R2SCORE)
```

```python
from sklearn.linear_model import Ridge, Lasso

# Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)
ridge_pred = ridge.predict(x_test)
print("Ridge R2 Score:", r2_score(y_test, ridge_pred)*100)
```

```python
# lasso regression 
lasso = Lasso(alpha=1.0)
lasso.fit(x_train,y_train)
lasso_predict = lasso.predict(x_test)
print("Lasso R2 score",r2_score(y_test,lasso_predict)*100)
```
